### Literal 2: 
Encuentra todos los puntos en los que la curva

$$
\left( \frac{y}{2} \right)^2 = \left( \frac{x + 3}{2} \right)^2 - 1
$$

interseca el eje \( y \) en \( y = -2 \).

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Funciones para Newton-Raphson
def f(x):
    return ((x + 3) / 2)**2 - 2

def f_prime(x):
    return (x + 3) / 2

# Rango de x
x_vals = np.linspace(-8, 2, 1000)
# Curva completa
y_pos = 2 * np.sqrt(((x_vals + 3) / 2)**2 - 1)
y_neg = -y_pos

# Aproximaciones iniciales
aproximaciones = [-6, 0]
raices = []
iteraciones_totales = []

# Ejecutar Newton y guardar trayectoria
for x0 in aproximaciones:
    x_current = x0
    trayectoria = []
    for i in range(20):
        trayectoria.append(x_current)
        x_new = x_current - f(x_current) / f_prime(x_current)
        if abs(x_new - x_current) < 1e-6:
            raices.append(x_new)
            break
        x_current = x_new
    iteraciones_totales.append(trayectoria)

raices_unicas = np.unique(np.round(raices, 5))

# ------------------ ANIMACIÃN: GRÃFICA GENERAL ------------------
max_iters = max(len(t) for t in iteraciones_totales)
for step in range(max_iters):
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_pos, 'blue', label='Curva (rama superior)')
    plt.plot(x_vals, y_neg, 'blue', label='Curva (rama inferior)')
    plt.axhline(-2, color='gray', linestyle='--', label='LÃ­nea y = -2')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.7)

    for trayectoria in iteraciones_totales:
        if step < len(trayectoria):
            plt.scatter(trayectoria[step], -2, color='red', s=80)

    plt.title('MÃ©todo de Newton - Vista general')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()
    display(plt.gcf())
    if step < max_iters - 1:
        clear_output(wait=True)
    plt.close()
    time.sleep(0.5)

# ------------------ ANIMACIÃN: VISTA AMPLIADA ------------------
for step in range(max_iters):
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_pos, 'blue')
    plt.plot(x_vals, y_neg, 'blue')
    plt.axhline(-2, color='gray', linestyle='--')

    for trayectoria in iteraciones_totales:
        if step < len(trayectoria):
            x = trayectoria[step]
            plt.scatter(x, -2, color='red', s=100)

    plt.xlim(-7, 1)
    plt.ylim(-3, -1)
    plt.title('MÃ©todo de Newton - Vista ampliada')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    display(plt.gcf())
    if step < max_iters - 1:
        clear_output(wait=True)
    plt.close()
    time.sleep(0.5)

# ------------------ GRÃFICAS FINALES FIJAS ------------------

# GrÃ¡fica final general
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_pos, 'blue', label='Curva (rama superior)')
plt.plot(x_vals, y_neg, 'blue', label='Curva (rama inferior)')
plt.axhline(-2, color='gray', linestyle='--', label='LÃ­nea y = -2')
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axvline(0, color='black', linestyle='--', linewidth=0.7)
plt.scatter(raices_unicas, [-2]*len(raices_unicas), color='red', s=100, label='RaÃ­ces')
plt.title('Resultado final - Vista general')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

# GrÃ¡fica final ampliada
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_pos, 'blue')
plt.plot(x_vals, y_neg, 'blue')
plt.axhline(-2, color='gray', linestyle='--')
plt.scatter(raices_unicas, [-2]*len(raices_unicas), color='red', s=100, label='RaÃ­ces')
plt.xlim(-7, 1)
plt.ylim(-3, -1)
plt.title('Resultado final - Vista ampliada')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()

# Resultado final
print(f"RaÃ­ces encontradas (intersecciones en y = -2): {raices_unicas}")
```
